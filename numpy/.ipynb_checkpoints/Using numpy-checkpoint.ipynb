##Introducing `numpy`

```
Roll array elements along a given axis. 
Elements that roll beyond the last position are re-introduced at the first.
```

````python
#!/usr/bin/env python2.7

from numpy import (roll, zeros)
import time

grid_shape = (512, 512)


def laplacian(grid):
    return roll(grid, +1, 0) + roll(grid, -1, 0) + \
        roll(grid, +1, 1) + roll(grid, -1, 1) - 4 * grid


def evolve(grid, dt, D=1):
    return grid + dt * D * laplacian(grid)


def run_experiment(num_iterations):
    grid = zeros(grid_shape)

    block_low = int(grid_shape[0] * .4)
    block_high = int(grid_shape[0] * .5)
    grid[block_low:block_high, block_low:block_high] = 0.005

    start = time.time()
    for i in range(num_iterations):
        grid = evolve(grid, 0.1)
    return time.time() - start

if __name__ == "__main__":
    run_experiment(500)
````

This code gives us a 40x speedup through lesser, vectorized instructions (this assumptions always needs to be checked though!) and fewer cache misses (accounts for most of the speedup):

##In-place operations

Memory allocation is even more costly than Cache misses, which are handled by highly specialized hardware. We can avoid unnecessary allocations by preallocating memory at the beginning of the code and making most functions __in-place__:

````python
#!/usr/bin/env python2.7

import numpy as np
import time

grid_shape = (512, 512)


def laplacian(grid, out):
    np.copyto(out, grid)
    out *= -4
    out += np.roll(grid, +1, 0)
    out += np.roll(grid, -1, 0)
    out += np.roll(grid, +1, 1)
    out += np.roll(grid, -1, 1)


def evolve(grid, dt, out, D=1):
    laplacian(grid, out)
    out *= D * dt
    out += grid


def run_experiment(num_iterations):
    scratch = np.zeros(grid_shape)  # preallocation
    grid = np.zeros(grid_shape)

    block_low = int(grid_shape[0] * .4)
    block_high = int(grid_shape[0] * .5)
    grid[block_low:block_high, block_low:block_high] = 0.005

    start = time.time()
    for i in range(num_iterations):
        evolve(grid, 0.1, scratch)
        grid, scratch = scratch, grid  # we need to swap for next iteration, this is quite cheap
    return time.time() - start

if __name__ == "__main__":
    run_experiment(500)
````

Make sure, the code stays readable! The above changes improve speed by 29%, mostly because of a reduced number of page faults.

Using the `id()` command is an easy way to track, whether two arrays reference the same physcal memory.

##Further improvements

Linewise profiling will tell us that `np.roll()` needs a lot of resources. This is because the function creates a new vector with every roll. We can replace this function by our own, specialized version.

A further improvement can be gained by using `numexpr`(https://github.com/pydata/numexpr), which optimizes algebraic operations on *large* arrays (for small arrays, however, the overhead outweights the speed improvement).

Both approaches are implemented in `diffusion_numpy_memory2_numexpr.py`:

````python
#!/usr/bin/env python2.7

from numpy import (copyto, multiply, zeros)
import numexpr as ne
import time

grid_shape = (512, 512)


def roll_add(rollee, shift, axis, out):
    if shift == 1 and axis == 0:
        out[1:, :] += rollee[:-1,:]
        out[0, :] += rollee[-1,:]
    elif shift == -1 and axis == 0:
        out[:-1, :] += rollee[1:,:]
        out[-1, :] += rollee[0,:]
    elif shift == 1 and axis == 1:
        out[:, 1:] += rollee[:, :-1]
        out[:, 0] += rollee[:, -1]
    elif shift == -1 and axis == 1:
        out[:, :-1] += rollee[:, 1:]
        out[:, -1] += rollee[:, 0]


def laplacian(grid, out):
    copyto(out, grid)
    multiply(out, -4.0, out)
    roll_add(grid, +1, 0, out)
    roll_add(grid, -1, 0, out)
    roll_add(grid, +1, 1, out)
    roll_add(grid, -1, 1, out)


def evolve(grid, dt, out, D=1):
    laplacian(grid, out)
    ne.evaluate("out*D*dt+grid", out=out)


def run_experiment(num_iterations):
    scratch = zeros(grid_shape)
    grid = zeros(grid_shape)

    block_low = int(grid_shape[0] * .4)
    block_high = int(grid_shape[0] * .5)
    grid[block_low:block_high, block_low:block_high] = 0.005

    start = time.time()
    for i in range(num_iterations):
        evolve(grid, 0.1, scratch)
        grid, scratch = scratch, grid
    return time.time() - start

if __name__ == "__main__":
    run_experiment(500)
````
